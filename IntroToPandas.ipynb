### Using loc

```python
#use loc to extract data using more textual conditions, instead of integer based location
fire = df.loc[df['Type 1'] == 'Fire']
```

```python
fire.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(fire.HP, fire.Attack, 'o')
plt.show()
```
